```python
%load_ext autoreload
%autoreload 2
from __future__ import print_function
from orphics import maps, cosmology, io, stats, lensing
import numpy as np
```

```python
pl = io.Plotter()
for minx in [1.0,0.1,0.01,0.001]:

    x = np.arange(0.01,5,0.01)
    kappa = lensing.gnfw(x)

    pl.add(x,kappa)
pl.done()
```

```python
pl = io.Plotter(yscale='log',xscale='log')

for minx in [1.0,0.1,0.01,0.001]:

    x = np.arange(0.01,5,0.01)
    kappa = lensing.gnfw(x)

    
    fkappa = np.abs(np.fft.fft(kappa))
    ls = np.fft.fftfreq(kappa.size, minx)*2*np.pi
    
    pl.add(ls[ls>0],fkappa[ls>0])
pl.done()
```
